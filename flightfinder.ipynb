```python
# import libraries
from bs4 import BeautifulSoup
from splinter import Browser
import pandas as pd
import datetime, time
from flask import Flask, jsonify
from pprint import pprint
```

```python
depdate = pd.to_datetime(input('Departure date in mm-dd-yyyy format'))
start_date = depdate.strftime('%Y-%m-%d')
start_date
```

```python
retdate = pd.to_datetime(input('Return date in mm-dd-yyyy format'))
return_date = retdate.strftime('%Y-%m-%d')
return_date
```

```python
#Define parameters


depcity = input('Departing from' )

```

```python
#Scraping data 
url ='https://www.kayak.com/flights/'+depcity+'-ORD/'+start_date+'/'+return_date+'?sort=bestflight_a'
browser = Browser('chrome')
browser.visit(url) 
time.sleep(15)
data = BeautifulSoup(browser.html, 'html.parser')
print(data)
```

```python
cheapest = data.find('div', class_='Hv20-value').text
cheapest
```

```python
airlines = []
c = 0
for air in data.findAll('div',{'class':'c_cgF'}):
    if c == 0:
        airlines.append(air.text.strip())
    if c % 5 == 0:
        airlines.append(air.text.strip())
        c = c + 1
    else:
        c = c + 1
        
#pprint(airlines)
           
```

```python
fr_airline = []
rt_airline = []
d = 0

for air in airlines:
    if d == 0:
        pass
    if d % 2 ==0:
        rt_airline.append(air)
        d = d + 1
    else:
        fr_airline.append(air)
        d = d + 1
#pprint(fr_airline)
#pprint(rt_airline)
```

```python
price=[]
for prices in data.findAll('div',{'class':'f8F1'}):
    price.append(prices.text.strip())
    
print(price)
```

```python
times = []
a = 0
for time in data.findAll('div',{'class':'vmXl'}):
    if a == 0:
        times.append(time.text.strip())
    if a % 3 == 0:
        times.append(time.text.strip())
        a = a + 1
    else:
        a = a + 1
#pprint(times)
```

```python
fr_time = []
rt_time = []
b = 0

for t in times:
    if b == 0:
        pass
    if b % 2 == 0:
        rt_time.append(t)
        b = b + 1
    else:
        fr_time.append(t)
        b = b + 1
del rt_time[0]
#pprint(fr_time)
#pprint(rt_time)
```

```python
tickets_df = pd.DataFrame(list(zip(fr_time, rt_time, fr_airline, price)), columns = ['Departure', 'Return','Airline','Price'])
tickets_df
```
