```python
import requests
import json
from pprint import pprint
from api_keys import geoapify_key
import pandas as pd
```

```python
df = pd.DataFrame(columns= ['Name', 'Lattitude', 'Longitude', 'Address'])

df.head()
```

```python

# Set the geographical coordinates
latitude = 41.9803
longitude = -87.9090

# Set the parameters for the type of search
categories = f"accommodation.hotel"
limit = 100
radius = 25000
filters = f"circle:{longitude},{latitude},{radius}"
bias = f"proximity:{longitude},{latitude}"

# set up a parameters dictionary
params ={
    "categories": categories,
    "filter":filters,
    "limit":limit,
    "bias":bias,
     "apiKey":geoapify_key}

     # Set base URL
url = "https://api.geoapify.com/v2/places"

#Make an API request using the params dictionary
response = requests.get(url, params=params).json()

response
```

```python
for hotels in response['features']:
    place_id = hotels['properties']['place_id']
    hotel_lat = hotels["properties"]["lat"]
    hotel_lon = hotels["properties"]["lon"]
    address = hotels["properties"]["address_line2"]
    hotel_name = hotels["properties"]["name"]

    df = df.append({'Name': hotel_name,
    'Lattitude': hotel_lat,
    'Longitude': hotel_lon,
    'Address': address
    }, ignore_index= True)
    

 
```

```python
df.to_csv('hotel_geoapfy_api_call.csv', index=True)
```
