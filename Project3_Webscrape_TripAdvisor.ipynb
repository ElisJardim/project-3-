```python
from splinter import Browser
from bs4 import BeautifulSoup
import requests
import pandas as pd
import string
from pprint import pprint
```

```python
browser = Browser('chrome')

url = 'https://www.tripadvisor.com/HotelsNear-g35805-qORD-Chicago_Illinois.html'

browser.visit(url)

html = browser.html

# Create a BeautifulSoup object from the scraped HTML
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Find and extract data elements.
hotels = []
for name in soup.findAll('div',{'class':'listing_title'}):
    hotels.append(name.text.strip())
ratings = []
for rating in soup.findAll('a',{'class':'ui_bubble_rating'}):
    ratings.append(rating['alt'])
reviews = []
for review in soup.findAll('a',{'class':'review_count'}):
    reviews.append(review.text.strip())
prices = []
for p in soup.findAll('div',{'class':'price-wrap'}):
    prices.append(p.text.replace('$','').strip())

hotels_clean = []
[hotels_clean.append(x) for x in hotels if x not in hotels_clean]

hotels_clean

keyword = 'Sponsored'
for i in hotels_clean[:]:
    if i.find(keyword) != -1:
        hotels_clean.remove(i)

len(hotels_clean)



```

```python
# Create the dictionary.
dict = {'Ratings':ratings,'Number of Reviews':reviews}

# Create the dataframe.
chicago = pd.DataFrame.from_dict(dict)
chicago.head(30)
```
